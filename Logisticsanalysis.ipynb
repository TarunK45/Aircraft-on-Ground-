```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
part_numbers = [
    '0002-AC108', 'ER-222', 'BZ-9876', 'AC-5678', 'DL-2309', 'TS-6541', 'MB-8901', 
    'FR-1287', 'RT-4329', 'LG-5478', 'HY-0987', 'NV-2345', 'FL-1209', 'EL-9876', 'WB-8765'
] * 10  

part_descriptions = [
    'Engine 1 Turbine', 'Landing Gear Front', 'Fuel Injector', 'Wing Flap Left', 
    'Navigation System', 'Hydraulic Pump', 'Electrical System Control', 
    'Avionics Processor', 'Spoilers Control Unit', 'Tail Fin Stabilizer', 
    'Fuel System', 'Rudder Control', 'Combustor Unit', 'Turbine Blade', 'Nose Cone Frame'
] * 10  

suppliers = [
    ('RTX Corporation', 'United States'), ('Airbus', 'Germany'), ('Lockheed Martin', 'United States'),
    ('Boeing', 'United States'), ('Northrop Grumman', 'United States'), ('Rostec', 'Russia'),
    ('BAE Systems', 'United Kingdom'), ('GE Aerospace', 'United States'), ('Safran', 'France'),
    ('Rolls-Royce Holdings', 'United Kingdom'), ('L3Harris', 'United States'), 
    ('Leonardo (Finmeccanica)', 'Italy'), ('United Technologies', 'United States'),
    ('Raytheon Company', 'United States')
]
```

```python
n = len(part_numbers)
data = {
    'Part Number': part_numbers,
    'Part Description': part_descriptions,
    'Supplier': np.random.choice([s[0] for s in suppliers], size=n),
    'Country': np.random.choice([s[1] for s in suppliers], size=n),
    'Quantity in Stock': np.random.randint(0, 50, size=n),
    'Cost per Part (USD)': np.random.randint(1000, 100000, size=n),
    'Time to Deliver (Hours)': np.random.randint(3, 36, size=n),  
    'Calibration Status': np.random.choice(['Checked', 'Unchecked'], size=n)
}
```

```python
df_supply_chain = pd.DataFrame(data)
```

```python
df_aircraft_parts = pd.read_csv('aircraft_parts_data_with_airlines_updated.csv')
```

```python
df_supply_chain['Aircraft Type'] = np.random.choice(df_aircraft_parts['Aircraft Type'], size=n)
df_supply_chain['Airline'] = np.random.choice(df_aircraft_parts['Airline'], size=n)
df_supply_chain['Aircraft Number'] = np.random.choice(df_aircraft_parts['Aircraft Number'], size=n)
```

```python
df_supply_chain.to_csv('aircraft_supply_chain_data.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_supply_chain = pd.read_csv('aircraft_supply_chain_data.csv')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Supplier', y='Quantity in Stock', data=df_supply_chain, errorbar=None)
plt.xticks(rotation=90)
plt.title('Stock Availability by Supplier')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Supplier', y='Time to Deliver (Hours)', data=df_supply_chain, errorbar=None)
plt.xticks(rotation=90)
plt.title('Average Delivery Time by Supplier')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Cost per Part (USD)', y='Time to Deliver (Hours)', data=df_supply_chain)
plt.title('Cost vs Delivery Time')
plt.show()
```
